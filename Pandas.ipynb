```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Amit', 'Raj', 'Sara', 'Pooja', 'John'],
    'Age': [25, 30, 22, 29, 35],
    'City': ['Delhi', 'Mumbai', 'Kolkata', 'Pune', 'Chennai']
}

df = pd.DataFrame(data)

# Filter rows where Age >20
filtered_df = df[df['Age'] >25]
print(filtered_df)
```

```python
df[(df['Age'] > 25) & (df['City'] == 'Mumbai')]
```

```python
df[(df['Age'] > 25) | (df['City'] == 'Delhi')]
```

```python
import pandas as pd
import numpy as np

# Sample DataFrame with NaN values
data = {
    'Name': ['Amit', 'Raj', 'Sara', 'Pooja', 'John'],
    'Age': [25, np.nan, 22, 29, np.nan],
    'City': ['Delhi', 'Mumbai', np.nan, 'Pune', 'Chennai']
}

df = pd.DataFrame(data)

# Check missing values
print(df.isna())  
df.fillna(0, inplace=True)
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.dropna(subset=['Name'], inplace=True)
df
```
