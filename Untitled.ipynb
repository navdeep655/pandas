```python
import pandas as pd

# Creating a DataFrame with NaN values
data = {
    'A': [1, 2, None, 4, None],
    'B': [None, 2, 3, None, 5],
    'C': [10, None, None, 40, 50]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)


fill=pd.Series({2:3,4:4})
fill2=pd.Series({0:1,3:4})
df['A'].fillna(fill,inplace=True)
df['B'].fillna(fill2,inplace=True)
df.replace({50:4,10:3},inplace=True)
df.isnull().sum().sum()
```

```python
df.replace([2,1],30,inplace=True)
df
```

```python
df['C'].fillna(df['C'].mean(),inplace=True)
df
```

```python
import pandas as pd

# Creating a DataFrame with duplicate rows
data = {
    'A': [1, 2, 2, 3, 4, 4, 5],
    'B': ['x', 'y', 'y', 'z', 'w', 'w', 'v']
}
df = pd.DataFrame(data)

df_cleaned = df.drop_duplicates()
print(df_cleaned)
```

```python
import pandas as pd

data = {
    'Student': ['Amit', 'Ravi', 'Neha', 'Rohit', 'Simran', 'Tina', 'Arjun', 'Riya'],
    'Class': ['10th', '10th', '10th', '11th', '11th', '12th', '12th', '12th'],
    'Marks': [85, 78, 92, 88, 76, 95, 82, 90],
    'salary':[2,3,4,4,4,5,5,4]
}

df = pd.DataFrame(data)
print(df.groupby('Class')[['Marks','salary']].apply(lambda x: x.max()-x.min()))
df.groupby('Class')['Marks'].apply(lambda x: pd.Series({
    'Range': x.max() - x.min(),
    'Min': x.min()
}))

```

```python
import pandas as pd

data = {
    'Class': ['10th', '10th', '10th', '11th', '11th', '12th', '12th', '12th'],
    'Gender': ['M', 'F', 'M', 'M', 'F', 'M', 'F', 'F'],
    'Marks': [85, 78, 92, 88, 76, 95, 82, 90],
    'Salary': [40000, 45000, 50000, 55000, 60000, 62000,55000,60000]
}

df = pd.DataFrame(data)
print(df)
```
