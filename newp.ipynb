```python
import numpy as np
import pandas as pd
df=pd.read_csv(r"C:\Users\maman\Downloads\Telegram Desktop\car_dataset.csv")
df
```

```python
df.drop("3",axis=1,inplace=True)
df
```

```python
df.drop("?",axis=1,inplace=True)
df
```

```python
df.columns = [ "Brand", "Fuel Type", "Aspiration", "Num of Doors", "Body Style", "Drive Wheels", "Engine Location", "Wheelbase", "Length", "Width", "Height", "Curb Weight", "Engine Type", "Num of Cylinders", "Engine Size", "Fuel System", "Bore", "Stroke", "Compression Ratio", "Horsepower","Peak RPM", "City_MPG", "Highway MPG", "Price"]
df
```

```python
df.isnull().sum()
```

```python
df.replace("?",np.nan,inplace=True)
df
```

```python
df["Num of Doors"].replace(np.nan,"two",inplace=True)
df
```

```python
df["Bore"] = df["Bore"].astype(float)
df["Bore"].fillna(df["Bore"].mean(),inplace=True)
df["Stroke"] = df["Stroke"].astype(float)
df["Stroke"].fillna(df["Stroke"].mean(),inplace=True)
df
```

```python
#print(df["Bore"].to_string())
df["Horsepower"].fillna(method="bfill",inplace=True)
df["Peak RPM"].fillna(method="bfill",inplace=True)
```

```python
df.dropna(subset=["Price"],inplace=True)
df
```
